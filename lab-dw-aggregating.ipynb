```python
yes_response = df[df["response"] == "Yes"]
yes_response.head()
```

```python
summary = yes_response.groupby(["gender", "policy_type"])[["monthly_premium_auto","customer_lifetime_value", "total_claim_amount"]].mean().round(2).reset_index()
summary.head()
```

```python
summary["profit_score"] = summary["customer_lifetime_value"] - summary["total_claim_amount"]
summary.head()
```

```python
#Check profitablity
summary_sorted = summary.sort_values(by="profit_score", ascending= False).reset_index(drop=True)
print(summary_sorted)
```

```python
#Low risk Segments
low_risk = summary.sort_values(by="total_claim_amount").reset_index(drop=True)
low_risk
```

```python
# your code goes here
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)
df.head()
```

```python
#Clean the columns with Lower case and Replace " " with "_"
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.head()
```

```python
Life_time = df.groupby(["education", "gender"]).agg(
    max_customer_value=("customer_lifetime_value", "max"),
    min_customer_value=("customer_lifetime_value", "min"),
    median_customer_value=("customer_lifetime_value", "median")
).round(2).reset_index()

Life_time.head()
```

```python
Life_time.sort_values(by="median_customer_value", ascending=False)
```
